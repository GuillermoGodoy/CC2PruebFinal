Para esta tarea se considerará un sistema de 2 masas conectadas por una cuerda. La primera masa en rojo será denotada por $p_1$ y es un péndulo. La segunda masa en verde se denota como $p_2$ y actúa como un peso, es decir solo movimiento vertical.

Ambas masas se ubican en planos de acción distintos, por lo cual no hay colisión entre ellas, y además el largo de la cuerda es fijo (2.5 metros). Inicialmente $p_1$ se ubica en el punto $\langle-1,0\rangle$ y $p_2$ en $\langle0.5,-0.3\rangle$, ámbos con velocidad inicial $0$.

A continuación se presenta una visualización de la situación.

```python
# You must install plotly in order to see the pendulum in action
with open('pendulum_data.json') as file:
    data = json.load(file)
    
figure1 = go.FigureWidget()

instant_line = go.Scatter()
instant_line.name = 'String'
instant_line.showlegend = False
instant_line.mode = 'lines'
instant_line.x = [data['x'][0], 0, 0.5, 0.5]
instant_line.y = [data['y'][0], 0, 0, data['r'][0]-1.3]
instant_line.line.width = 2
instant_line.line.color = 'blue'
figure1.add_trace(instant_line)

instant1 = go.Scatter()
instant1.name = 'P1'
instant1.mode = 'markers'
instant1.x = [data['x'][0]]
instant1.y = [data['y'][0]]
instant1.marker.size = 15
instant1.marker.color = 'red'
figure1.add_trace(instant1)

instant2 = go.Scatter()
instant2.name = 'P2'
instant2.mode = 'markers'
instant2.x = [0.5]
instant2.y = [data['r'][0]-1.3]
instant2.marker.size = 15
instant2.marker.color = 'green'
figure1.add_trace(instant2)

figure1.layout.hovermode = 'closest'
figure1.layout.height = 560
figure1.layout.title = 'CC2 - IVP Pendulum'

figure1.layout.yaxis.scaleanchor = 'x'
figure1.layout.yaxis.scaleratio = 1

figure1.layout.xaxis.range = [-1.2, 1.2]
figure1.layout.yaxis.range = [-1.2, 0.5]

slider = w.IntSlider(min=0, max=len(data['x']), step=1)

def move_pend(update):
    with figure1.batch_animate(duration=0):
        weights1 = figure1.data[1]
        weights1.x = [data['x'][update['new']]]
        weights1.y = [data['y'][update['new']]]
        weights2 = figure1.data[2]
        weights2.y = [data['r'][update['new']]-1.3]
        line = figure1.data[0]
        line.x = [data['x'][update['new']], 0, 0.5, 0.5]
        line.y = [data['y'][update['new']], 0, 0, data['r'][update['new']]-1.3]
    
slider.observe(move_pend, names='value')

play = w.Play(min=0, max=len(data['x']), step=10, interval=1)

w.jslink((play, 'value'), (slider, 'value'))

w.VBox([
    w.HBox([play, slider]),
    figure1
])
```

```python
# Forward Euler Method
def euler_ode(y,t,f,h):
    return y+h*f(t,y)

# Runge-Kutta of Second order
def RK2_ode(y,t,f,h):
    k1=y+h/2.0*f(t,y) #or euler_ode(y,t,f,h)
    return y+h*f(t+h/2.0,k1)

# Runge-Kutta 
def RK4_ode(y,t,f,h):
    k1=f(t,y)
    k2=f(t+h/2.0,y+(h/2.0)*k1)
    k3=f(t+h/2.0,y+(h/2.0)*k2)
    k4=f(t+h,y+h*k3)
    return y+(h/6.0)*(k1+2.0*k2+2.0*k3+k4)
```

### 1.- Llegando a las ecuaciones [30 puntos] 
---

Modele la trayectoria de $p_1$ y $p_2$ utilizando coordenadas polares $\theta$ y $r$. Para el ángulo, deberá tomar en cuenta que es un péndulo de largo variable, mientras que para el largo del péndulo basta con utilizar la segunda ley de Newton tomando en cuenta los pesos y la fuerza centrípeta. 
Considere que ambas masas tienen el mismo volumen, sin embargo la densidad de $p_2$ es $5$ veces mayor a la de $p_1$.
El pivote de $p_1$ es la coordenada $\langle0,0\rangle$ y el pivote de $p_2$ es $\langle0.5,0\rangle$. La unidad del sistema son metros, es decir, la coordenada $\langle0.5,0\rangle$ se encuentra a 50 cm del origen.

## Desarrollo

Realicemos en un comienzo el estudio por separado, veamos en primer lugar que pasa con $p_1$, para ello considere la siguiente figura esquematica:
<img src="dibujo_p1.png"></img>
Considere ahora una representación polar del problema, como se muestra a continuación:
<img src="fuerzas.jpg"></img>

Note que en este esquema, la masa experimenta dos fuerzas, saber la fuerza peso $m_1g$ y la fuerza de tensión que genera la cuerda, dado el enfoque polar descomponemos la fuerza peso en las compontes $m_1g\sin(\theta)$ y $m_1g\cos(\theta)$, sobre el arco el objeto posee una aceleración tangencial $a$ por lo que la ecuación correspondiente sería:
\begin{equation}
-m_1g\sin(\theta(t))=m_1a_{t}(t)
\end{equation}

Pero recordemos que la acelaración tangencial en un movimiento circular, sigue la relación:

\begin{equation}
a_{t}(t)=l(t)\ddot{\theta}(t)
\end{equation}

De donde reemplazando en lo anterior se sigue que:

$$l(t)\ddot{\theta}(t)+\sin(\theta(t))=0$$

Por otra parte note que dado que la cuerda tiene largo variable, la masa posee movimiento y aceleración vértical, por lo que haciendo el balance en este contexto se obtiene que:

$$m_1g\cos(\theta(t))-N=m_1\ddot{l}(t)$$

Con $N$ la fuerza de tensión.Pensemos ahora en el esquema de $p_2$
<img src="dibujo_p2.png"></img>
En este caso solo existe movimiento vertical, y llamemos $y(t)$ a la altura de $p_2$ en el tiempo y para ser consistente con lo hecho para $p_1$ considere que $y(t)$ aumenta hacia arriba, luego por la segunda ley de Newton se tiene que:

$$T-m_2g=m_2a_{p2}(t)$$

Pero note lo siguente $a_{p2}(t)=\ddot{y}(t)$, como la cuerda tiene un largo fijo $L$ se sigue que $y(t)=L-l(t)$ pues cuando $p_1$ este con la mayor cantidad de cuerda más abajo $p_2$ estará más arriba y por el sistema de referencia usado tendrá su mayor valor, por otra parte considereando que no hay perdidas en la cuerda las tensiones $N$ y $T$ deben ser iguales,de esta forma se obtiene que:

$$N-m_2g=-m_2\ddot{l}(t)$$

Finalmente despejando y juntando la información se obtiene el siguiente set de ecuaciones.

\begin{equation}
 \begin{matrix}
  l(t)\ddot{\theta}(t)+\sin(\theta(t))=0 \\
  (m_1-m_2)\ddot{l}(t)+g(m_2-m_1\cos(\theta(t))=0
 \end{matrix}
\end{equation}

### 2.- Estudio numérico 1 [40 puntos] 
---

Tomando las ecuaciones obtenidas en la pregunta anterior, **escriba el modelo** como sistema dinámico y realice un **Análisis de Estabilidad Lineal**. En base a los resultados del análisis **seleccione un método adecuado**, **justificando** claramente su elección. **Grafique** el comportamiento del sistema a través del tiempo.

Considere las siguientes variables auxiliares $h(t)=\dot{l}(t)$ y $\gamma(t)=\dot{\theta}(t)$, y defina $X(t)=(l(t),h(t),\theta(t),\gamma(t))^{T}$, y note que del set de ecuaciones obtenido en la pregunta 1 se sigue que:

$$\dot{X}(t)= \dot{\left(\begin{array}{l}
l(t) \\
h(t) \\
\theta(t) \\
\gamma(t) 
\end{array} \right)}=F(l(t),h(t),\theta(t),\gamma(t))=\left(\begin{array}{l}
\ \ \ \ \ \ \ h(t) \\
-g\frac{(m_2-m_1\cos(\theta(t)))}{m_1-m_2} \\
\ \ \ \ \ \ \ \gamma(t) \\
\ \ \  -\frac{\sin(\theta(t))}{l(t)}
\end{array} \right)
$$

Note que lo anterior se puede ver como $\dot{X}(t)=F(X(t))$ lo que corresponde claramente a un sistema dinámico, pero con la salvedad que de $F$ tiene caracter No-Lineal

Dada la caracteristica no lineal del sistema, para hacer un analisis de estabilidad lineal debemos estudiar la matriz Jacobiana del sistema.Calculando obtenemos que:

$$DF(l(t),h(t),\theta(t),\gamma(t))=\left( \begin{array}{llll}
0 & 1 & \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ 0 & 0 \\
0 & 0 & -g \cdot m_1\sin(\theta(t))\cdot\gamma(t) & 0\\
0 & 0 & \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ 0 & 1 \\
\frac{\sin(\theta(t))}{l^{2}(t)}& 0 & \ \ \ \ \ -\frac{\cos(\theta(t))}{l(t)} &0 
\end{array} \right) $$

Se procede a cargar el sistema y su matriz jacobiana

```python
g=scipy.constants.g
m1=10
m2=5*m1
Large=2

def ode_system(t,y):
    y1 = y[0]
    y2 = y[1]
    y3 = y[2]
    y4 = y[3]
    return np.array([y2, -g*(m2-m1*np.cos(y3))/(m1-m2),y4,-np.sin(y3)/y1])
    #return np.array([y2, g*(-m2+m1*np.cos(y3))/(m2-m1),y4,-np.sin(y3)/(Large-y1)])
def my_vdp_eig_jacobian(t,y):
    J=[[0,1,0,0],[0,0,-g*m1*np.sin(y[2])*y[3]]/(m1-m2),[0,0,0,1],[np.sin(y[2])/(y[0]^2),0,-1*np.cos(y[2])/y[0],0,0]]
    lambs,vs= LA.eig(J)
    return lambs
```

```python
y0=[5, 1,0.1,-1]
L=10
h=0.01

T=20

t_times = np.arange(0, T, h)
y_output = np.zeros([t_times.size,4])
y_output[0,:] = y0
for i in range(1,t_times.size):
    y_output[i,:]=euler_ode(y_output[i-1,:],t_times[i-1],ode_system,h)
    
fig = pyplot.figure(figsize=(L,L))
ax = fig.gca()
ax.grid(True)
ax.set_title("Numerical Approximation axis_x p_2")
plot(t_times,np.multiply(np.cos(y_output[:,2]),y_output[:,0]),'-')

fig = pyplot.figure(figsize=(L,L))
ax = fig.gca()
ax.grid(True)
ax.set_title("Numerical Approximation axis_y p_2")
plot(t_times,np.multiply(np.sin(y_output[:,2]),y_output[:,0]),'-')


fig = pyplot.figure(figsize=(L,L))
ax = fig.gca()
ax.grid(True)
ax.set_title("Phase Portrait")
plot(np.multiply(np.cos(y_output[:,2]),y_output[:,0]),np.multiply(np.sin(y_output[:,2]),y_output[:,0]),'-')
plt.xlabel('axis_x',fontsize=24);
plt.ylabel('axis_y',fontsize=24);
```

Probamos con el método de Euler las ecuaciones propuestas y vemos que estás no modelan el real comportamiento. Una explicación de esto puede ser la falta de conociemiento técnico en física por parte del desarrollador.

De igual manera, en los gráficos obtenidos puede verse que la coordenada $x$  e $y$ del pendulo divergen oscilando, lo que también se observa en el retrato de fase 

Tras lo anterior buscaremos estableces las ecuaciones que modelan el fénomeno mediante un argumento de *Mécanica Lagrangeana*. Para ello estudiemos en primer lugar las ecuaciones que gobiernas a $p_1$, llame $K$ y $V$ a la energía cinética y potencial del pendulo respectivamente, luego se tiene que:

$$K=\frac{1}{2}\left( \dot{l}^2+(l\dot{\theta})^2\right)$$

$$V=-m_1gl\cos(\theta)$$

Luego el lagrangeano del sistema viene dado por:

$$L=K-V=\frac{1}{2}\left( \dot{l}^2+(l\dot{\theta})^2\right)+m_1gl\cos(\theta)$$

Planteando las ecuaciones de *Euler-Lagrange* $\dfrac{d}{dt}\left(\dfrac{\partial L}{\partial \dot{\theta}} \right)-\dfrac{\partial L}{\partial \theta}=0$  se tiene que:

$$\begin{array}{lcl}
\text{Para } \ \ \ \theta & \implies &  l^2\ddot{\theta}+2l\dot{\theta}+gl\sin(\theta)=0 \\
\text{Para } \ \ \ l  &\implies &\ddot{l}-l\dot{\theta}^{2}-g\cos(\theta)=0 
\end{array}$$

Para la particula $p_2$ considere un sistema de refencia donde la coordenada $y$ aumente hacia arriba, note que como está solo posee movimiento vertical se tiene que:

$$T-m_2g=m_2\ddot{y}$$

Donde $T$ representa la tensión que siente la masa $p_2$ hacia arriba e $y(t)$ representa la posición de la particula $p_2$ en el plano.

Note que la tensión, suponiendo una cuerda ideal debe ser igual a lo largo de toda la cuerda y tome $T_5=m_1g\left(1+2\frac{h_m}{\bar{L}}\right)$ con $\bar{L}=L-0,5$ el largo máximo que puede tener la cuerda en el pendulo y en la casa que sube y baja y $h_m=l(t)\cos(\theta(t))$ finalmente como $T=T_5$ se sigue que:

$$l^2\ddot{\theta}+2l\dot{\theta}+gl\sin(\theta)=0$$

$$\ddot{l}-l\dot{\theta}^{2}-g\cos(\theta)=0 $$

$$m_1g\left(1+l(t)\cos(\theta(t))\right)- m_2g=m_2\ddot{y}$$
 
Pues $L=2,5$

Considere ahora las siguientes variables auxiliares para escribir lo anterior como un sistema dinámico, $x=\dot{y}$, $h=\dot{l}$ y $\gamma=\dot{\theta}$,luego el sistema anterior se ve como:

$$\dot{X}(t)= \dot{\left(\begin{array}{l}
y(t) \\
x(t) \\
l(t) \\
h(t) \\
\theta(t) \\
\gamma(t) 
\end{array} \right)}=F(y(t),x(t),l(t),h(t),\theta(t),\gamma(t))=\left(\begin{array}{l}
\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ x(t) \\
\frac{1}{m_2}\left(m_1g\left(1+l(t)\cos(\theta(t))\right)- m_2g\right) \\
\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ h(t) \\
\ \ \ \ \ \ \ \ \ \ \ \ \  l{\gamma}^{2}+g\cos(\theta) \\
\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \gamma(t) \\
\ \ \ \ \ \ \ \ \ \ \ \ \  -\frac{2l\gamma+gl\sin(\theta)}{l^2}
\end{array} \right)
$$

```python
g=scipy.constants.g
m1=10
m2=5*m1
Large=2

def new_ode_system(t,y):
    y1 = y[0]
    y2 = y[1]
    y3 = y[2]
    y4 = y[3]
    y5 = y[4]
    y6 = y[5]
    return np.array([y2, (m1*g*(1+y3*np.cos(y5))-m2*g)/m2,y4,y3*y6*y6+g*np.cos(y5),y6,-(2*y3*y6+g*y3*np.sin(y5))/(y3*y3)])
    #return np.array([y2, g*(-m2+m1*np.cos(y3))/(m2-m1),y4,-np.sin(y3)/(Large-y1)])
def my_vdp_eig_jacobian(t,y):
    J=[[0,1,0,0],[0,0,-g*m1*np.sin(y[2])*y[3]]/(m1-m2),[0,0,0,1],[np.sin(y[2])/(y[0]^2),0,-1*np.cos(y[2])/y[0],0,0]]
    lambs,vs= LA.eig(J)
    return lambs
```

```python
y0=[-0.3,0,1.7,0,-90,0]
L=10
h=0.001

T=10

t_times = np.arange(0, T, h)
y_output = np.zeros([t_times.size,6])
y_output[0,:] = y0
for i in range(1,t_times.size):
    y_output[i,:]=euler_ode(y_output[i-1,:],t_times[i-1],new_ode_system,h)
   
    
fig = pyplot.figure(figsize=(L,L))
ax = fig.gca()
ax.grid(True)
ax.set_title("Numerical Approximation axis_x p_2")
plot(t_times,np.multiply(np.cos(y_output[:,2]),y_output[:,0]),'-')

fig = pyplot.figure(figsize=(L,L))
ax = fig.gca()
ax.grid(True)
ax.set_title("Numerical Approximation axis_y p_2")
plot(t_times,np.multiply(np.sin(y_output[:,2]),y_output[:,0]),'-')


fig = pyplot.figure(figsize=(L,L))
ax = fig.gca()
ax.grid(True)
ax.set_title("Phase Portrait")
plot(np.multiply(np.cos(y_output[:,5]),y_output[:,3]),np.multiply(np.sin(y_output[:,5]),y_output[:,3]),'-')
plt.xlabel('axis_x',fontsize=24);
plt.ylabel('axis_y',fontsize=24);
```
